```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('DataSet/pokemon.csv', index_col=0)
```

```python
df.head()
```

```python
sns.lmplot(x='Attack', y='Defense', data=df)
```

```python
sns.lmplot(x='Attack', y='Defense', 
           fit_reg=False,
           hue='Generation',
           data=df)
```

```python
sns.lmplot(x='Attack', y='Defense', 
           fit_reg=False,
           hue='Generation',
           data=df)
plt.ylim(0, None)
plt.xlim(0, None)
```

```python
sns.boxplot(data=df)
```

```python
stats_df = df.drop(['Total', 'Generation', 'Legendary'], axis=1)
sns.boxplot(data=stats_df, notch =True, linewidth = 2.5, width = 0.50)
```

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

```python
sns.violinplot(x='Type 1', y='Attack', data = df, 
               palette = pkmn_type_colors)
```

```python
sns.set_style('whitegrid')
sns.violinplot(x='Type 1', y='Attack', data = df)
```

```python
sns.swarmplot(x='Type 1', y='Attack', data = df, palette=pkmn_type_colors)
```

```python
sns.violinplot(x='Type 1', y='Attack', data = df,
              inner = None, palette=pkmn_type_colors)
sns.swarmplot(x='Type 1', y='Attack', data = df, 
              color = 'k', alpha = 0.7)
plt.title('Attack by Type')
```

```python
sns.swarmplot(x='Stat', y='value', data = melted_df, hue='Type 1')
```

```python
plt.figure(figsize=(10,6))

sns.swarmplot(x='Stat',
              y='value',
              data=melted_df,
              hue='Type 1',
              split=True,
              pallette=pkmn_type_colors)
plt.ylim(0,260)
plt.legend(bbox_to_anchor=(1,1), loc=2)
```
